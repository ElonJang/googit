```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.figure(figsize=(12,6))
plt.plot(x, y);
```

```python
plt.figure(figsize=(12,6))

plt.plot(x, y, 'k', label='Original')
plt.plot(x, np.cos(x), 'b', label='True Cos')
plt.plot(x, np.r_[0, np.diff(y)]/0.1, 'r.', label='Calculated Diff')

plt.legend(loc='best')
plt.show()
```
